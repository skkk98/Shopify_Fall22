```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
data = pd.read_csv(r'C:\Users\kalya\Downloads\ShopifyData.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
# we can see that the order_amount mean/avg is 3145.13 which is very high for a sneaker shope
# This is probably because we are not considering the total_items!!!!
# So we try to calculate the Average Order Value(AOV) after calculating order amount based on 1 item!

order_amount_per_one = data
order_amount_per_one["order_amount_per_one"] = data["order_amount"]/data["total_items"]
order_amount_per_one["shop_id"] = data["shop_id"]
```

```python
order_amount_per_one.head()
```

```python
new_aov = sum(order_amount_per_one["order_amount_per_one"])/len(order_amount_per_one["order_amount_per_one"])
print(new_aov)
```

```python
# We can see that the Average Order value is $387.74. This makes a little more sense, Although it's still high for an affordable sneaker shoes.
# So we'll try to calculate the same AOV but based on each shop.
shops = data["shop_id"].unique()
print(shops)
```

```python
for shop in shops:
    print("Average Order Value(AOV) of Shop "+str(shop)+" is $"+str(np.mean(order_amount_per_one["order_amount_per_one"][order_amount_per_one["shop_id"]==shop])))
```
